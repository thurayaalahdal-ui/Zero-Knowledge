```python

import random
import hashlib

# القيم العامة
p = 23
x = 5                     # السر
v = pow(x, 2, p)          # المفتاح العام v = x^2 mod p

def verify(t, e, s):
    return pow(s, 2, p) == (t * pow(v, e, p)) % p

# ---------------- Honest ----------------
r = random.randint(1, p-1)
t = pow(r, 2, p)
e = random.randint(0, 1)
s = (r * pow(x, e, p)) % p
print("--- Honest ---")
print("t=", t, "e=", e, "s=", s, "Verified=", verify(t, e, s))

# ---------------- Cheat ----------------
# الغشاش ما يعرف السر x، فيخمن فقط
fake_r = random.randint(1, p-1)
t = pow(fake_r, 2, p)
e = random.randint(0, 1)
# يحاول يغطي على نفسه لكنه ما يعرف x
s = fake_r                  # غلط متعمد
print("\n--- Cheat ---")
print("t=", t, "e=", e, "s=", s, "Verified=", verify(t, e, s))

# ---------------- Fiat–Shamir (Non-interactive) ----------------
r = random.randint(1, p-1)
t = pow(r, 2, p)
# التحدي مشتق من Hash (غير تفاعلي)
h = hashlib.sha256(str(t).encode()).hexdigest()
e = int(h, 16) % 2
s = (r * pow(x, e, p)) % p

print("\n--- Fiat–Shamir ---")
print("t=", t, "e=", e, "s=", s, "Verified=", verify(t, e, s))

# ---------------- Cheating Probability ----------------
trials = 100
fails = 0
for _ in range(trials):
    fake_r = random.randint(1, p-1)
    t = pow(fake_r, 2, p)
    e = random.randint(0, 1)
    s = fake_r
    if verify(t, e, s):
        fails += 1

print("\nCheating success rate =", fails / trials)
```
