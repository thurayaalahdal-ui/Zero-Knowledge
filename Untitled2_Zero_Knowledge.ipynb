```python
import random, hashlib

p=23; g=5; x=6; y=pow(g,x,p)
r=random.randint(1,p-1)
t=pow(g,r,p)
e=random.randint(0,1)
s=(r+e*x)%p
print("--- Honest ---")
print("t=",t,"e=",e,"s=",s,"Verified=",pow(g,s,p)==(t*pow(y,e,p))%p)
t_f=random.randint(1,p-1)
e=random.randint(0,1)
s_f=random.randint(1,p-1)
print("\n--- Cheat ---")
print("t=",t_f,"e=",e,"s=",s_f,"Verified=",pow(g,s_f,p)==(t_f*pow(y,e,p))%p)
r=random.randint(1,p-1)
t=pow(g,r,p)
e_fs=int(hashlib.sha256(str(t).encode()).hexdigest(),16)%2
s=(r+e_fs*x)%p
print("\n--- Fiat–Shamir ---")
print("t=",t,"e=",e_fs,"s=",s,"Verified=",pow(g,s,p)==(t*pow(y,e_fs,p))%p)
success=0
for _ in range(100):
    t_f=random.randint(1,p-1)
    e=random.randint(0,1)
    s_f=random.randint(1,p-1)
    if pow(g,s_f,p)==(t_f*pow(y,e,p))%p:
        success+=1
print("\nCheating probability =",success/100)
```
